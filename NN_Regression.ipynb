## Neural Network 
INput layer size = 5 
2 hidden layers each with 100 nodes ( Relu Activation function )
Output layer with single unit ( summation function)
Regression problem 
Learning rate = 0.001
No of epochs = 300 ( no of times we pass the entire data set through network)
Adam Solver 
Loss / cost function =mean square error ( MSE)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from matplotlib import pyplot as plt
```

```python
re = pd.read_excel(r'/Users/yusufkhan/Desktop/data analytics/deep learning/Real estate valuation data set.xlsx')
re.shape
```

```python
re.head()
```

```python
X = re.iloc[:,2:-1]
Y = re.iloc[:,-1]
Y = np.ravel(Y)
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(Xtrain)
Xtrain = scaler.transform(Xtrain)
Xtest = scaler.transform (Xtest)
```

```python
NN = MLPRegressor (max_iter =300, activation ='relu', hidden_layer_sizes = (100,100))
```

```python
MLPRegressor
```

```python
# Fitting the model 
NN.fit(Xtrain,Ytrain)
```

```python
# Predict 
NN_pred = NN.predict(Xtest)
```

```python
print("MSE:",mean_squared_error(Ytest,NN_pred))
print("MAE:",mean_absolute_error(Ytest,NN_pred))
```

```python
print("R2 Score:",r2_score(Ytest,NN_pred))
```
